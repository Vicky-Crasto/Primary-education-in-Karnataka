```r
#Subsetting the variable needed for analysis
colnames(data1)
school_data <- data1[,c(1:18,27:29,33,34,36:52,55,58)]

colnames(data2)
school_enroll_data <- data2 [, c(1:3,9,14,20,25,33,37,43,47,53,58,64,69,70:88,98)]

colnames(data4)
census_data <- data4[,c(1:15,27,35,38:40)]
```

```r
#Creating variable to understand the data

#Percentage of childern in the age group of 6 to 15 
school_data$perc_low_prim_pop <- round((school_data$TOT_6_10_15/school_data$TOTPOPULAT)*100,2)
school_data$perc_upp_prim_pop <- round((school_data$TOT_11_13_15/school_data$TOTPOPULAT)*100,2)

#Percentage of single classroom school
school_data$prec_sing_class <- round((school_data$SCLSTOT/school_data$SCHTOT)*100,2)

#Percentage of single teacher school
school_data$prec_sing_teach <- round((school_data$STCHTOT/school_data$SCHTOT)*100,2)

#Percentage of Schools Approachable by All Weather Road
school_data$prec_approach_road <- round((school_data$ROADTOT/school_data$SCHTOT)*100,2)

#Percentage of schools with Playground Facility
school_data$prec_playground <- round((school_data$SPLAYTOT/school_data$SCHTOT)*100,2)

# Percentage of Schools with Boundarywall
school_data$prec_boundarywall <- round((school_data$SBNDRTOT/school_data$SCHTOT)*100,2)

# Percentage of Schools with Girls' Toilet
school_data$prec_girl_toilet <- round((school_data$SGTOILTOT/school_data$SCHTOT)*100,2)

# Percentage of Schools with Boys' Toilet
school_data$prec_boy_toilet <- round((school_data$SBTOILTOT/school_data$SCHTOT)*100,2)

#Percentage of Schools with Drinking Water
school_data$prec_drinking_water <- round((school_data$SWATTOT/school_data$SCHTOT)*100,2)

#Percentage of Schools with Electricity
school_data$prec_electricity <- round((school_data$SELETOT/school_data$SCHTOT)*100,2)

#Percentage of schools with computers
school_data$prec_computer <- round((school_data$SCOMPTOT/school_data$SCHTOT)*100,2)

# Percentage of schools with Schools with Ramp (where needed)
school_data$prec_ramp <- round((school_data$SRAMTOT/school_data$SCHTOT)*100,2)

# Percentage of schools providing mid meal (govt aid)
school_data$prec_midday_meal <- round((school_data$MDMTOT/school_data$SCHTOT)*100,2)

# Percentage of Schools with Kitchen-Shed (Government & Aided Schools)
school_data$prec_kitchen_shed <- round((school_data$KITTOT/school_data$SCHTOT)*100,2)

# Percentage of Schools where Mid-Day Meal is Provided and Prepared in School Premises (Government & Aided Schools)_Total
school_data$prec_prep_serve_sch <- round((school_data$KITSTOT/school_data$SCHTOT)*100,2)

# Ratio of teacher to student
school_data$total_enrollment <- school_data$TOT_6_10_15 + school_data$TOT_11_13_15
school_data$teacher_student_ratio <-round((school_data$ENRTOT/school_data$TCHTOT))

# Percentage of girl enrollemnt to total school enrollment
school_data$girl_child_enroll <-round((school_data$ENRGTOT/school_data$total_enrollment)*100,2)
```

```r
#Combining enrollment data from School data with school_enroll_data
#create a dataset with district and total enrollment

enroll_data <- school_data[,c("DISTNAME", "ENRTOT")]
school_enroll_data <- left_join(school_enroll_data,enroll_data,by = "DISTNAME")
```

```r
# Total school enrollment in each district

school_enroll_data$total_enroll_sum <- school_enroll_data$C1_5_B + school_enroll_data$C6_9_B+school_enroll_data$C1_5_G+
  school_enroll_data$C6_9_G +school_enroll_data$C1_5_BD+school_enroll_data$C6_8_BD+school_enroll_data$C1_5_GD + school_enroll_data$C1_8_GD

#Perc of enroll for boys grade 1 to 5
school_enroll_data$prec_enroll_boys_1_5 <- round((school_enroll_data$C1_5_B/school_enroll_data$total_enroll_sum)*100,2)
#Perc of enroll for boys grade 6 to 9
school_enroll_data$prec_enroll_boys_6_9 <- round((school_enroll_data$C6_9_B/school_enroll_data$total_enroll_sum)*100,2)

#Perc of enroll for girls grade 1 to 5
school_enroll_data$prec_enroll_girls_1_5 <- round((school_enroll_data$C1_5_G/school_enroll_data$total_enroll_sum)*100,2)
#Perc of enroll for girls grade 6 to 9
school_enroll_data$prec_enroll_girls_6_9 <- round((school_enroll_data$C6_9_G/school_enroll_data$total_enroll_sum)*100,2)

#Perc of enroll for special need boys grade 1 to 5
school_enroll_data$prec_enroll_spboys_1_5 <- round((school_enroll_data$C1_5_BD/school_enroll_data$total_enroll_sum)*100,2)
#Perc of enroll for special boys grade 6 to 8
school_enroll_data$prec_enroll_spboys_6_8 <- round((school_enroll_data$C6_8_BD/school_enroll_data$total_enroll_sum)*100,2)

#Perc of enroll for sp girls grade 1 to 5
school_enroll_data$prec_enroll_spgirls_1_5 <- round((school_enroll_data$C1_5_GD/school_enroll_data$total_enroll_sum)*100,2)
#Perc of enroll for sp girls grade 6 to 9
school_enroll_data$prec_enroll_spgirls_6_8 <- round((school_enroll_data$C1_8_GD/school_enroll_data$total_enroll_sum)*100,2)

# Perc of enroll of SC girls in primary
school_enroll_data$perc_scgirls_pri <- round((school_enroll_data$SCPTOT_G/school_enroll_data$SCPTOT)*100,2)

# Perc of enroll of SC girls in upper primary
school_enroll_data$perc_scgirls_uppri <- round((school_enroll_data$SCUTOT_G/school_enroll_data$SCUTOT)*100,2)


# Perc of enroll of ST girls in primary
school_enroll_data$perc_stgirls_pri <- round((school_enroll_data$STPTOT_G/school_enroll_data$STPTOT)*100,2)

# Prec of enroll of ST girls in upper primary
school_enroll_data$perc_stgirls_uppri <- round((school_enroll_data$STUTOT_G/school_enroll_data$STUTOT)*100,2)

# Prec of enroll of muslim girls in primary
school_enroll_data$perc_mugirls_pri <- round((school_enroll_data$MUUTOT/school_enroll_data$MUPTOT)*100,2)

# Prec of enroll of muslim girls in upper primary
school_enroll_data$perc_mugirls_uppri <- round((school_enroll_data$MUUTOT_G/school_enroll_data$MUPTOT_G)*100,2)

# Prec of class in good condition
school_enroll_data$class_good_cond <- round((school_enroll_data$TOTCLMITOTAL/school_enroll_data$TOTCLSTOTAL)*100,2)

# Prec of classes needing minor repair
school_enroll_data$class_minor_rep <- round((school_enroll_data$TOTCLGDTOTAL/school_enroll_data$TOTCLSTOTAL)*100,2)

# Prec of classes needing major repair
school_enroll_data$class_major_rep <- round((school_enroll_data$TOTCLMJTOTAL/school_enroll_data$TOTCLSTOTAL)*100,2)
```

```r
#Using census data
#Perc of literate
census_data$perc_literate <- round((census_data$Literate/census_data$Population)*100,2)

#Perc of male literate
census_data$perc_male_literate <- round((census_data$Male_Literate/census_data$Population)*100,2)

#Perc of female literate
census_data$perc_female_literate <- round((census_data$Female_Literate/census_data$Population)*100,2)

#Perc of SC population
census_data$perc_sc_popul <- round((census_data$SC/census_data$Population)*100,2)

#Perc of ST population
census_data$perc_st_popul <- round((census_data$ST/census_data$Population)*100,2)

#Perc of Muslim population
census_data$perc_musl_popul <- round((census_data$Muslims/census_data$Population)*100,2)

# Perc of urban household
census_data$perc_urban_household <- round((census_data$Urban_Households/census_data$Households)*100,2)

# Perc of rural household
census_data$perc_rural_household <- round((census_data$Rural_Households/census_data$Households)*100,2)

# Perc of household with electricity
census_data$perc_electri_household <- round((census_data$Housholds_with_Electric_Lighting/census_data$Households)*100,2)
```

```r
# Combining the needed statistic in one dataset

school_data_ext <- school_data[,c(1:5,43:61)]
school_enroll_ext <- school_enroll_data[,c(3,38:54)]
census_data_ext <- census_data[,c(3,4,20:29)]

primary_edu_data <- left_join(school_data_ext,school_enroll_ext,by ="DISTNAME" )
primary_edu_data1 <- left_join(primary_edu_data,census_data_ext,by=c("DISTNAME"="District.name"))
```

```r
#Creating districtwise scorecard basis school infrastructure

school_infra <- primary_edu_data1[,c(3,10:20,39:41)]

school_infra$approach_road <- school_infra$prec_approach_road/100
school_infra$playground <- school_infra$prec_playground/100
school_infra$boundarywall <- school_infra$prec_boundarywall/100
school_infra$girls_toilet <- school_infra$prec_girl_toilet/100
school_infra$boys_toilet <- school_infra$prec_boy_toilet/100
school_infra$drinking_water <- school_infra$prec_drinking_water/100
school_infra$electricity <- school_infra$prec_electricity/100
school_infra$computer <- school_infra$prec_computer/100
school_infra$ramp <- school_infra$prec_midday_meal/100
school_infra$kitchen_shed <- school_infra$prec_kitchen_shed/100
school_infra$classroom_cond <- (((school_infra$class_good_cond*0.60)+(school_infra$class_minor_rep*0.30)+(school_infra$class_major_rep*0.10))/100)

school_infra%>%select(16:26)%>%rowSums()->school_infra$total_score
```

```r
#creating a scorecard district wise basis focus on girl child education
girl_education <-primary_edu_data1[,c(3,24,27,28,31:38)]

girl_education$overall_enrollment <- girl_education$girl_child_enroll/100
girl_education$primary_enrollment <- girl_education$prec_enroll_girls_1_5/100
girl_education$upper_prim_enroll <- girl_education$prec_enroll_girls_6_9/100
girl_education$spec_prim_enroll <- girl_education$prec_enroll_spgirls_1_5/100
girl_education$spec_upper_prim_enroll <- girl_education$prec_enroll_spgirls_6_8/100
girl_education$SC_primary <- girl_education$perc_scgirls_pri/100
girl_education$SC_upper_prim <- girl_education$perc_scgirls_uppri/100
girl_education$ST_primary <- girl_education$perc_stgirls_pri/100
girl_education$ST_upper_prim <- girl_education$perc_stgirls_uppri/100
girl_education$muslim_prim <- girl_education$perc_mugirls_pri/100
girl_education$muslim_upper_prim <- girl_education$perc_mugirls_uppri/100

girl_education%>%select(13:23)%>%rowSums()->girl_education$total_score

head(girl_education)
```

```r
#Districtwise score card based on overal education quality and demographic feature
education_qual <- primary_edu_data1[,c(3,8,9,23,44,45,46,49)]

education_qual$Single_class <- education_qual$prec_sing_class/(-100)
education_qual$single_teacher <- education_qual$prec_sing_teach/(-100)
education_qual$pupil_teacher_ratio <- as.numeric(as.character(cut(education_qual$teacher_student_ratio,breaks =c(-Inf,30,35,40,Inf),labels= c(1,0.75,0.50,0.25))))

education_qual$literate_pop <- education_qual$perc_literate/(100)
education_qual$male_literate_pop <- education_qual$perc_male_literate/(100)
education_qual$female_literate_pop <- education_qual$perc_female_literate/(100)

education_qual%>%select(9:14)%>%rowSums()->education_qual$total_score

head(education_qual)
```
